```python
import urllib.request
import csv
res = "https://datacenter.taichung.gov.tw/swagger/OpenData/6721533d-55c4-4379-9bd4-b4cd35935189"
urllib.request.urlretrieve(res, './data/old.csv')
```

```python
import pandas as pd 
df = pd.read_csv('./data/old.csv')
```

```python
df1 = df[['行政區','辦理單位']].groupby(['行政區']).count()
df1
```

```python
#人口資料
res = "https://datacenter.taichung.gov.tw/swagger/OpenData/fdb8ae40-2513-4edf-9e7e-472890ab3e7a"
urllib.request.urlretrieve(res, './data/people.csv')
people = pd.read_csv('./data/people.csv')
people = people.groupby(['區別']).sum()
people = people.T[65:].sum()
people = pd.DataFrame(people)
people.columns = ['人口']
```

```python
df2 = df1.merge(people,how = 'outer',left_index = True, right_index = True)
df2['人口單位比'] = df2['辦理單位'] / df2['人口']
df2
```
